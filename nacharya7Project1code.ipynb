# Classifiers on  Pokémon Dataset


To start off my analysis of different algorithms, I plan to categorize Pokémon into either Legendary or not based on the features provided in the dataset.

```python
#Imports for running script
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn import preprocessing
from IPython.display import Image as PImage
from subprocess import check_call

```

# POKEMON DATA SET TEST

Testing the viability of a Legendary Pokemon Classifier.

```python
dataframe = pd.read_csv("./data/Pokemon.csv")

df = dataframe.copy()

## Change the TYpe1 Type2 Columns into a singlur feature.

df["Types"] = df['Type 1'] + df['Type 2']
df.loc[df["Types"].isnull(), "Types"] = df["Type 1"]

 #Tokenize Types
type_encoder = preprocessing.LabelEncoder()
type_encoder.fit(df['Types'].unique().astype(str))
df['Types'] = type_encoder.transform(df['Types'].astype(str))


#drop columns we don't care about

drop_elements = ["#", "Name", "Type 1", "Type 2", "Generation"]
df = df.drop(drop_elements, axis=1)
dfcopy = df.copy()

from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size=0.625)
print(train.head())

metric = "f1"

y = dfcopy["Legendary"]
x = dfcopy.drop(["Legendary"], axis=1) 

train_y = train["Legendary"]
train_x = train.drop(["Legendary"], axis=1)

test_y = test["Legendary"]
test_x = test.drop(["Legendary"], axis=1)
```

## Decision Tree Results

```python
from sklearn import tree

parameters = {'max_depth':range(1,9)}
clf = GridSearchCV(tree.DecisionTreeClassifier(), parameters, cv=5, scoring=metric)
clf.fit(X=train_x, y=train_y)

tree_model = clf.best_estimator_
print (clf.best_score_, clf.best_params_) 

y_pred = tree_model.predict(test_x)

print("Testing set accuracy: ", f1_score(test_y, y_pred, pos_label=True))
```

## Neural Network Performance

```python
import warnings
warnings.filterwarnings('ignore')

## For the neural network, we will use more examples in training, going for an 80/20 split instead of the 3/5 split 
## we used in the decision tree since the learning curve showed more improvement with more samples.

train_nn, test_nn = train_test_split(df, test_size=0.2)

train_y_nn = train_nn["Legendary"]
train_x_nn = train_nn.drop(["Legendary"], axis=1)

test_y_nn = test_nn["Legendary"]
test_x_nn = test_nn.drop(["Legendary"], axis=1)

print()

from sklearn.neural_network import MLPClassifier

param_grid={
#     'learning_rate': [0.05, 0.01, 0.005, 0.001],
    'activation': ["relu", "tanh"],
    'solver': ['lbfgs'],
    'hidden_layer_sizes': [(x,y) for x in range(1,30) for y in [1,2]],
    'max_iter': [200 ,400, 600, 800, 1000]}

nn_clf = GridSearchCV(MLPClassifier(), param_grid=param_grid, scoring=metric, cv=3)
nn_clf.fit(X=train_x_nn, y=train_y_nn)

nn_model = nn_clf.best_estimator_
print(nn_clf.best_score_, nn_clf.best_params_)

y_pred = nn_model.predict(test_x_nn)
print("Testing set accuracy: ", f1_score(test_y_nn, y_pred))
```

## AdaBoost Learner

```python
from sklearn.ensemble import AdaBoostClassifier

dtc = tree.DecisionTreeClassifier(max_depth=clf.best_params_["max_depth"])
abe = AdaBoostClassifier(base_estimator=dtc)

param_grid = {
    "n_estimators": [x for x in range(1,50)] 
}

abe_clf = GridSearchCV(AdaBoostClassifier(), param_grid=param_grid, scoring=metric, cv=5)

abe_clf.fit(X=train_x, y=train_y)

abe_model = abe_clf.best_estimator_
print(abe_clf.best_score_, abe_clf.best_params_)

y_pred = abe_model.predict(test_x)
print("Testing set accuracy: ", f1_score(test_y, y_pred))
```

## Support Vector Machine

```python
from sklearn.svm import SVC
from sklearn.model_selection import StratifiedKFold

param_grid = {'C': [1], 'kernel': ['linear', 'rbf']}

svm_clf = GridSearchCV(SVC(), param_grid=param_grid, scoring=metric, cv=5)

svm_clf.fit(X=train_x, y=train_y.astype(int))

svm_model = svm_clf.best_estimator_
print(svm_clf.best_score_, svm_clf.best_params_)

y_pred = svm_model.predict(test_x)
print("Testing set accuracy: ", f1_score(test_y, y_pred))

```

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier

param_grid = {
    "n_neighbors": range(1,50),
    "weights": ["uniform", "distance"]
}

knn_clf = GridSearchCV(KNeighborsClassifier(), param_grid=param_grid, scoring=metric, cv=5)

knn_clf.fit(X=train_x, y=train_y)

knn_model = knn_clf.best_estimator_

print(knn_clf.best_score_, knn_clf.best_params_)

y_pred = knn_model.predict(test_x)
print("Testing set accuracy: ", f1_score(test_y, y_pred))
```

# Letter Handwriting Classifier

Similar to MNIST, this dataset contains a bunch of metadata on images of handwritten letters for classification. Instead of actually having the pixel matrix, however, we are just given 16 different measurements to base the classification on.
Dataset found [here.](https://archive.ics.uci.edu/ml/machine-learning-databases/letter-recognition/letter-recognition.data)

```python
names=['lettr','xbox','ybox','width','high','onpix','xbar','ybar','x2bar','y2bar','xybar','x2ybr','xy2br','xege','xegvy','yege','yegvx']
dataframe = pd.read_csv('./data/letter-recognition.csv',header=None, names=names, index_col=False)

df = dataframe.copy()

#Encode the letter labels to integers, one-hot style

classes = df['lettr'].unique()

letter_encoder = preprocessing.LabelEncoder()
letter_encoder.fit(df['lettr'].unique().astype(str))
df['lettr'] = letter_encoder.transform(df['lettr'].astype(str))

#Split data into train and test in a 4:1 split, according to the source documentation
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size=0.2)
print(train.head())

metric = "accuracy"

def score(model, x, y, metric):
    if metric == "accuracy":
        return model.score(x,y)
    elif metric == "f1":
        y_pred = model.predict(x)
        return f1_score(y, y_pred)
    elif metric == "f1_macro":
        y_pred = model.predict(x)
        return f1_score(y, y_pred, average="macro")

dfcopy = df.copy()
y = dfcopy["lettr"]
x = dfcopy.drop(["lettr"], axis=1)

train_y = train["lettr"]
train_x = train.drop(["lettr"], axis=1)

test_y = test["lettr"]
test_x = test.drop(["lettr"], axis=1)
```

## Decision Tree Performance

```python
from sklearn import tree

parameters = {'max_depth': [16],
              'criterion': ['entropy'],
              'min_samples_leaf': range(1,10)}
clf = GridSearchCV(tree.DecisionTreeClassifier(), parameters, cv=5, scoring=metric)
clf.fit(X=train_x, y=train_y)

tree_model = clf.best_estimator_
print (clf.best_score_, clf.best_params_)

print("Testing set accuracy: ", score(tree_model, test_x, test_y, metric))
```

## Neural Network Performance

```python
from sklearn.neural_network import MLPClassifier

param_grid={
    'activation': ["relu"],
    'solver': ['adam'],
    'hidden_layer_sizes': [(x,y) for x in range(40,60, 5) for y in [1,2]],
    'max_iter': [200, 400, 600, 800, 1000]}

nn_clf = GridSearchCV(MLPClassifier(), param_grid=param_grid, scoring=metric, cv=3)
nn_clf.fit(X=train_x, y=train_y)

nn_model = nn_clf.best_estimator_
print(nn_clf.best_score_, nn_clf.best_params_)

print("Testing set accuracy: ", score(nn_model, test_x, test_y, metric))
```

## Support Vector Machine Performance

```python
from sklearn.svm import SVC

param_grid = {'C': [100], 'kernel': ['rbf', 'linear']}

svm_clf = GridSearchCV(SVC(), param_grid=param_grid, scoring=metric, cv=5)

svm_clf.fit(X=train_x, y=train_y)

svm_model = svm_clf.best_estimator_
print(svm_clf.best_score_, svm_clf.best_params_)

y_pred = svm_model.predict(test_x)
print("Testing set accuracy: ", score(svm_model, test_x, test_y, metric))
```

## K Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier

param_grid = {
    "n_neighbors": range(1,11),
    "weights": ["uniform", "distance"]
}

knn_clf = GridSearchCV(KNeighborsClassifier(), param_grid=param_grid, scoring=metric, cv=5)

knn_clf.fit(X=train_x, y=train_y)

knn_model = knn_clf.best_estimator_

print(knn_clf.best_score_, knn_clf.best_params_)

y_pred = knn_model.predict(test_x)
print("Testing set accuracy: ", score(knn_model,test_x,test_y, metric))
```
